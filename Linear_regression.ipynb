```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error,r2_score
```

```python
d=pd.read_csv("/content/AmesHousing.csv")
d
x=d.drop('SalePrice',axis=1)
y=d['SalePrice']
n=x.select_dtypes(include=['int64','float64']).columns
c=x.select_dtypes(include=['object']).columns
n_t=Pipeline(steps=[('s',SimpleImputer(strategy='median')),('sc',StandardScaler())])
n_t
c_t=Pipeline(steps=[('d',SimpleImputer(strategy='most_frequent')),('sd',OneHotEncoder(handle_unknown='ignore'))])
c_t
```

# New Section

```python
pre=ColumnTransformer(transformers=[('n',n_t,n),('c',c_t,c)])
pre
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
m=Pipeline(steps=[('p',pre),('lr',LinearRegression())])
m
```

```python
m.fit(x_tr,y_tr)
y_pr=m.predict(x_te)
```

```python
print("acc",r2_score(y_te,y_pr)*100)
```
